### Agent Name: Academic Advisor Agent
### Agent Type: Simple Reflex Agent
### Objective: Reads GPA and attendance, then recommends an action using IF-THEN rules.

```python
class StudentAdvisorAgent:
    def __init__(self):
        self.gpa = 0.0
        self.attendance = 0.0

    def perceive(self, input_gpa, input_attendance):
        """
        Perception Layer:
        The agent perceives data from the environment and stores it.
        """
        self.gpa = input_gpa
        self.attendance = input_attendance

    def act(self):
        """
        Reasoning Layer:
        The agent applies IF-THEN rules to the perceived data to decide an action.
        """
        # Rule 1: High GPA and High Attendance
        if self.gpa >= 3.5 and self.attendance >= 90:
            return "Excellent! Recommend for University Scholarship."
            
        # Rule 2: Normal GPA and Good Attendance
        elif self.gpa >= 2.0 and self.attendance >= 75:
            return "Good standing. Allowed to register for next semester."
            
        # Rule 3: Low GPA but Good Attendance
        elif self.gpa < 2.0 and self.attendance >= 75:
            return "Academic Warning. Focus on improving your GPA."
            
        # Rule 4: Normal/High GPA but Bad Attendance
        elif self.gpa >= 2.0 and self.attendance < 75:
            return "Attendance Warning. Potential risk of failure due to absence."
            
        # Rule 5: Low GPA and Bad Attendance
        elif self.gpa < 2.0 and self.attendance < 75:
            return "CRITICAL! At risk of academic dismissal. Meet advisor immediately."
            
        else:
            return "Please verify your input data."
```

### Environment Interaction (Perception & Action)


```python
# Environment Interaction
print("--- Welcome to the Smart Student Advisor Agent ---")

advisor = StudentAdvisorAgent()

try:
    # Get inputs from the environment
    user_gpa = float(input("Enter Student GPA (0.0 to 4.0): "))
    user_attendance = float(input("Enter Attendance Percentage (0 to 100): "))

    # Validation
    if not (0.0 <= user_gpa <= 4.0 and 0 <= user_attendance <= 100):
        print("Error: Values out of valid range.")
        exit()

    # Agent perceives the inputs
    advisor.perceive(user_gpa, user_attendance)
    
    # Agent decides the action based on its logic
    decision = advisor.act()
    
    # Output the final processing report for the user
    print("--- Agent Processing Report ---")
    print(f"Recorded GPA: {advisor.gpa}")
    print(f"Recorded Attendance: {advisor.attendance}%")
    print("--------------------------------")
    print(f"Agent Action >> {decision}")
    print("--------------------------------")

except ValueError:
    # Error handling for non-numeric inputs
    print("Error: Invalid input. Please enter numeric values only.")
```
